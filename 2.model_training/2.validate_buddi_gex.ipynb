### Load sc Pseudobulk

```python
pattern = re.compile(r'^(.*)_((?:Train)|(?:Test))_((?:meta)|(?:prop)|(?:pseudo))_splits\.pkl$')

# Temporary dictionary to group by sample_id
temp_dict = defaultdict(lambda: defaultdict(dict))

for path in pseudobulk_files:
    file = path.name
    match = pattern.match(file)
    if match:
        sample_id, datasplit, datatype = match.groups()
        temp_dict[datasplit][sample_id][datatype] = path

# Organizing the sorted results
grouped_files = {
    "Train": {"meta": [], "prop": [], "pseudo": []},
    "Test": {"meta": [], "prop": [], "pseudo": []}
}

# Sort within each datasplit by sample_id and organize the lists
for datasplit in ["Train", "Test"]:
    sorted_samples = sorted(temp_dict[datasplit].keys())  # Sort by sample_id
    for sample_id in sorted_samples:
        for datatype in ["meta", "prop", "pseudo"]:
            if datatype in temp_dict[datasplit][sample_id]:
                grouped_files[datasplit][datatype].append(temp_dict[datasplit][sample_id][datatype])

meta_sc_train = pd.concat(
    [pd.read_pickle(file) for file in grouped_files['Train']['meta']]
)
meta_sc_train[STIM_COL] = 'white' # assume all single cell samples are from white individuals
Y_sc_train = pd.concat(
    [pd.read_pickle(file) for file in grouped_files['Train']['prop']]
)
X_sc_train = pd.concat(
    [pd.read_pickle(file).loc[:, sc_bulk_intersection_genes] for file in grouped_files['Train']['pseudo']]
)
```

### Concatenate X, Y, Gene Feature Selection,  and Encode Metadata

```python
# Generate place holder for bulk proportion (not used)
Y_bulk_dummy = pd.DataFrame(
    np.zeros((X_bulk_train.shape[0], Y_sc_train.shape[1])),
    columns=Y_sc_train.columns
)

X_concat = pd.concat([X_bulk_train, X_sc_train])
Y_concat = pd.concat([Y_bulk_dummy, Y_sc_train])
meta_concat = pd.concat([meta_bulk_train, meta_sc_train])

# save gene and cell type names
X_gene_names = X_concat.columns.to_list()
Y_cell_type_names = Y_concat.columns.to_list()

## get the top variable genes
X_colmean = X_concat.values.mean(axis=0) # mean across samples
X_colvar = X_concat.values.var(axis=0) # variance across samples
# coefficient of variation which is the var to mean ratio
X_CoV = np.array(np.divide(X_colvar, X_colmean+0.001)) 

# need to get the genes such that
# the union of the highly variable and the
# CIBERSORTx genes are 7000 total
num_genes_found = False

gene_df = pd.DataFrame(X_gene_names, columns=['gene'])

# start with top 7000 genes by CoV
initial_count = 7000
while not num_genes_found:
    # retrieve the top initial_count genes by CoV
    idx_top = np.argpartition(X_CoV, -initial_count)[-initial_count:]
    # get gene names from idx_top
    top_gene_df = gene_df.iloc[idx_top]

    # produce the union of the top genes and the cibersort genes
    CoV_only = np.union1d(top_gene_df, cibersortx_sig_genes)

    # check if the union is 7000
    if len(CoV_only) == 7000:
        num_genes_found = True
    else:
        # if not decrement the top CoV gene to retrieve by 1
        # in the next iteration, there will be two possibilities
        # 1. the union will have one less gene due to the removed gene being only in the CoV genes
        # 2. the union will have the same number of genes as this iteration due to the removed 
        #    gene being also present in CIBERSORTx genes
        # keep running this loop until the union has 7000 genes
        initial_count = initial_count -1

idx_top = np.argpartition(X_CoV, -initial_count)[-initial_count:] # num_genes to get 7000
gene_df = gene_df.iloc[idx_top]
venn2([set(gene_df.values.flatten()), 
       set(cibersortx_sig_genes)], 
       set_labels = ('Top CoV Genes', 'Cibersort Genes'))
plt.show()

feature_select_genes = np.union1d(gene_df, cibersortx_sig_genes)

# to numpy matrix
X = X_concat.loc[:,feature_select_genes].to_numpy()
print(X.shape)

Y = Y_concat.to_numpy()

FIELDS_TO_ENCODE = [SAMPLE_ID_COL, STIM_COL, TECH_COL]

encoded = {}

for field in FIELDS_TO_ENCODE:
    encoder = OneHotEncoder(dtype=int)
    encoded[field] = pd.DataFrame(
        encoder.fit_transform(meta_concat.loc[:,[field]]).toarray(),
        columns=encoder.get_feature_names_out([field])
    )
```

```python
import numpy as np
import pandas as pd

def sample_by_category(z_mu, z_logvar, meta, column, n_samples=None, random_state=None):
    """
    Samples from z ~ N(mu, sigma^2) for each category in `meta[column]`.

    Args:
        z_mu (np.ndarray): Mean tensor of shape (n_samples_total, latent_dim)
        z_logvar (np.ndarray): Log variance tensor of shape (n_samples_total, latent_dim)
        meta (pd.DataFrame): DataFrame with metadata including the column used for splitting
        column (str): Column name in `meta` used for categorizing samples
        n_samples (int, optional): Number of random samples to draw per category.
                                   If None, use all available samples in each category.
        random_state (int, optional): Random seed for reproducibility

    Returns:
        dict: A dictionary where keys are category values and values are arrays
              of shape (n_samples, latent_dim)
    """

    z_mu = z_mu.numpy()
    z_logvar = z_logvar.numpy()

    assert len(z_mu) == len(meta), "Mismatch between z_mu rows and meta rows"
    assert z_mu.shape == z_logvar.shape, "z_mu and z_logvar must have the same shape"

    if random_state is not None:
        rng = np.random.default_rng(random_state)
    else:
        rng = np.random.default_rng()

    sampled_dict = {}

    for category in meta[column].unique():
        idx = meta[meta[column] == category].index.to_list()

        mu_cat = z_mu[idx,:]
        logvar_cat = z_logvar[idx,:]
        std_cat = np.exp(0.5 * logvar_cat)
        eps = rng.standard_normal(std_cat.shape)
        z_sampled = mu_cat + eps * std_cat

        sampled_dict[category] = z_sampled

    return sampled_dict

def sample_latent(z_mu, z_logvar, n_samples=1, random_state=None):
    """
    Samples from z ~ N(mu, sigma^2) using the reparameterization trick.

    Args:
        z_mu (tf.Tensor): Tensor of shape (batch_size, latent_dim)
        z_logvar (tf.Tensor): Tensor of shape (batch_size, latent_dim)
        n_samples (int): Number of samples per input to draw
        random_state (int, optional): Seed for reproducibility

    Returns:
        tf.Tensor: Sampled latent tensor of shape (batch_size, n_samples, latent_dim)
    """
    batch_size, latent_dim = tf.shape(z_mu)[0], tf.shape(z_mu)[1]
    std = tf.exp(0.5 * z_logvar)

    eps = tf.random.normal(shape=(batch_size, n_samples, latent_dim), seed=random_state)
    
    # Expand z_mu and std to (batch_size, n_samples, latent_dim) for broadcasting
    z_mu_exp = tf.expand_dims(z_mu, axis=1)
    std_exp = tf.expand_dims(std, axis=1)

    return z_mu_exp + eps * std_exp
```

```python
z_label_dict = sample_by_category(z_label_mu, z_label_logvar, meta_sc, 'sample_id', n_samples=10, random_state=None)
z_stim_dict = sample_by_category(z_stim_mu, z_stim_logvar, meta_sc, 'sample_id', n_samples=10, random_state=None)
z_samp_type_dict = sample_by_category(z_samp_type_mu, z_samp_type_logvar, meta_sc, 'sample_id', n_samples=10, random_state=None)
z_slack_dict = sample_by_category(z_slack_mu, z_slack_logvar, meta_sc, 'sample_id', n_samples=10, random_state=None)
```

```python
n = 100
rng = np.random.default_rng()

ls_impute_list = []
meta_impute = pd.DataFrame()

for sample_id in meta_sc['sample_id'].unique():
    for i in range(n_y):
        # 1. Generate one-hot dummy of shape (n, n_y)
        y_impute = np.zeros((n, n_y))
        y_impute[:, i] = 1

        # 2. Sample with replacement n rows from each latent dict
        z_label_sample = z_label_dict[sample_id]
        z_stim_sample = z_stim_dict[sample_id]
        z_samp_type_sample = z_samp_type_dict[sample_id]
        z_slack_sample = z_slack_dict[sample_id]

        idx_label = rng.choice(z_label_sample.shape[0], size=n, replace=True)
        idx_stim = rng.choice(z_stim_sample.shape[0], size=n, replace=True)
        idx_samp_type = rng.choice(z_samp_type_sample.shape[0], size=n, replace=True)
        idx_slack = rng.choice(z_slack_sample.shape[0], size=n, replace=True)

        sampled_label = z_label_sample[idx_label]
        sampled_stim = z_stim_sample[idx_stim]
        sampled_samp_type = z_samp_type_sample[idx_samp_type]
        sampled_slack = z_slack_sample[idx_slack]

        ls_impute_list.append(
            np.concatenate(
                [y_impute, sampled_label, sampled_stim, sampled_samp_type, sampled_slack], axis=1
            )
        )

        meta_impute = pd.concat([
            meta_impute,
            pd.DataFrame(data={
                'sample_id': [sample_id] * n,
                'cell_type': [i] * n
            })
        ])

ls_impute = np.concatenate(ls_impute_list, axis=0)

print(ls_impute.shape)
print(meta_impute.shape)
```
